# Problem 4

```python
def F(x, y):
    return y**(1/3)
```

```python
fig = plt.figure(figsize=(10,6))
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
xStop = 20
x1 = np.linspace(0, xStop, 1000)

x, y = integrate(F, 0, 0, xStop, 0.01)
axes.plot(x, y, label='initial condition y = 0')

x, y = integrate(F, 0, 1e-16, xStop, 0.01)
axes.plot(x, y, label='initial condition y = 1e-16')

axes.legend(loc=2,fontsize=14)
axes.set_xlabel(r"$x$",fontsize=18)
axes.set_ylabel(r"$y$",fontsize=18)
axes.set_title('solns')
```

# Problem 8

```python
def F(x, y):
    g = 9.80665
    cd = 0.2028
    m = 80
    F1 = y[1]
    F2 = g - (cd / m)*((y[1])**2)
    return np.array([F1, F2])
```

```python
y0 = np.array([-5000, 0])

x, y = integrate(F, 0, y0, 95, 0.01)

fig = plt.figure(figsize=(10,6))
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y[:, 0])

i = 0
while(y[i, 0] < 0):
    i += 1

for k in range(i-3, i+3):
    print(x[k], y[k])
```
