```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Foreward difference approximation

```python
def fp(h):
    return (np.around(np.sin(0.8+h), 5) - np.around(np.sin(0.8), 5))/h

def error(h):
    return np.abs(fp(h) - np.cos(0.8))
```

```python
h = np.linspace(0.0001, 0.006, 100000)
e = error(h)
optimalh = round(h[np.argmin(e)], 5)
print('Let h =', optimalh, '--> f\'(0.8) = ', round(fp(optimalh), 5))
```

```python
fig = plt.figure(figsize=(10,6))
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.set_xlabel(r"$h$",fontsize=18)
axes.set_ylabel(r"$error$",fontsize=18)
axes.set_title('error')
axes.plot(h, e)
```

# Central differerence approximation

```python
def fp(h):
    return ((np.around(np.sin(0.8+h), 5) - np.around(np.sin(0.8-h), 5))/(2*h))

def error(h):
    return np.abs(fp(h) - np.cos(0.8))
```

```python
h = np.linspace(0.001, 0.04, 10000000)
e = error(h)
optimalh = round(h[np.argmin(e)], 5)
print('Let h =', optimalh, '--> f\'(0.8) = ', round(fp(optimalh), 5))
```

```python
fig = plt.figure(figsize=(10,6))
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.set_xlabel(r"$h$",fontsize=18)
axes.set_ylabel(r"$error$",fontsize=18)
axes.set_title('error')
axes.plot(h, e)
```
