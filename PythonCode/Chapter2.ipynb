## Problem 25

```python
theta = np.pi / 4.0
g = 9.82
m = np.array([10., 4., 5., 6.])
mu = np.array([0.25, 0.3, 0.2])

a = np.array([[1., 0., 0., m[0]],
            [-1., 1., 0., m[1]],
            [0., -1., 1., m[2]],
            [0., 0., -1., m[3]]])
b = np.array([m[0]*g*(np.sin(theta) - mu[0]*np.cos(theta)),
             m[1]*g*(np.sin(theta) - mu[1]*np.cos(theta)),
             m[2]*g*(np.sin(theta) - mu[2]*np.cos(theta)),
             -m[3]*g])
x = gaussPivot(a, b)
print(x)
```
